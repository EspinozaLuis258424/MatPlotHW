```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv("data/mouse_drug_data.csv")
clinical_data = pd.read_csv("data/clinicaltrial_data.csv")

# Combine the data into a single dataset
combined_data = pd.merge(clinical_data,mouse_data, on = 'Mouse ID',how='outer')

# Display the data table for preview
combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_volume = combined_data.groupby(['Drug','Timepoint']).mean()['Tumor Volume (mm3)']

# Convert to DataFrame
tumor_mean_df = pd.DataFrame(tumor_volume.reset_index())
#reset_index() took me a while to figure out
# Preview DataFrame
tumor_mean_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem = combined_data.groupby(['Drug','Timepoint']).sem()['Tumor Volume (mm3)']

# Convert to DataFrame
tumor_sem_df = pd.DataFrame(tumor_sem.reset_index())
# Preview DataFrame
tumor_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_pivot = tumor_mean_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
tumor_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_pivot.index, tumor_pivot['Capomulin'],yerr=tumor_sem['Capomulin'],marker='o', color='r')
plt.errorbar(tumor_pivot.index, tumor_pivot['Placebo'],yerr=tumor_sem['Placebo'],marker='d',color='k')
plt.errorbar(tumor_pivot.index, tumor_pivot['Ketapril'],yerr=tumor_sem['Ketapril'],marker='s',color='g')
plt.errorbar(tumor_pivot.index, tumor_pivot['Infubinol'],yerr=tumor_sem['Infubinol'],marker='^',color='b')


plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend(loc='best')
plt.grid()
# Save the Figure

plt.show()
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
met_mean = combined_data.groupby(['Drug','Timepoint']).mean()['Metastatic Sites']

# Convert to DataFrame
met_mean_df = pd.DataFrame(met_mean.reset_index())

# Preview DataFrame
met_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sem = combined_data.groupby(['Drug','Timepoint']).sem()['Metastatic Sites']

# Convert to DataFrame
met_sem_df = pd.DataFrame(met_sem.reset_index())
# Preview DataFrame
met_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_pivot = met_mean_df.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')

# Preview that Reformatting worked
met_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(met_pivot.index, met_pivot['Capomulin'],yerr=met_sem['Capomulin'],marker='o', color='r')
plt.errorbar(met_pivot.index, met_pivot['Placebo'],yerr=met_sem['Placebo'],marker='d',color='k')
plt.errorbar(met_pivot.index, met_pivot['Ketapril'],yerr=met_sem['Ketapril'],marker='s',color='g')
plt.errorbar(met_pivot.index, met_pivot['Infubinol'],yerr=met_sem['Infubinol'],marker='^',color='b')


plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Met. Sites')
plt.legend(loc='best')
plt.grid()
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = combined_data.groupby(['Drug','Timepoint']).count()['Mouse ID']

# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count.reset_index())
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_pivot = mice_count_df.pivot(index='Timepoint',columns='Drug',values='Mouse ID')
# Preview the Data Frame
mice_pivot
```

```python
# Generate the Plot (Accounting for percentages)
for row in mice_pivot:
    mice_pivot[row]*4
mice_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mice_pivot.index, mice_pivot['Capomulin']/25*100, marker='o', color='r')
plt.errorbar(mice_pivot.index, mice_pivot['Placebo']/25*100, marker='d',color='k')
plt.errorbar(mice_pivot.index, mice_pivot['Ketapril']/25*100, marker='s',color='g')
plt.errorbar(mice_pivot.index, mice_pivot['Infubinol']/25*100, marker='^',color='b')


plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')
plt.legend(loc='best')
plt.grid()
# Save the Figure

# Show the Figure
plt.show()
```
